```r
mongo_uri <- "mongodb://localhost:27017/anvisa"
```

```r
processar_arquivo <- function(caminhoArquivo) {
  
  base_dados <- fread(caminhoArquivo, heade=TRUE, encoding="Latin-1")
  base_dados[base_dados==""] <- NA
  base_dados <- base_dados[complete.cases(base_dados), ]
  
  conn <- mongo("vendas_medicamentos_controlados_antimicrobianos", url = mongo_uri)
  conn$insert(base_dados)
  conn$disconnect()
  
}
```

```r
lapply(arquivos_entrada, processar_arquivo)
```
