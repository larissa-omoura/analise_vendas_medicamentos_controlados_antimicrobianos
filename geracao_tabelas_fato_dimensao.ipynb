```r
library(ggplot2)
```

```r
df_filtrado <- subset(fato_vendas_medicamentos_controlados_antimicrobianos, ANO_VENDA %in% c(2014, 2015))
```

```r
df_filtrado$DATA_VENDA <- with(df_filtrado, as.Date(paste(ANO_VENDA, MES_VENDA, "01", sep = "-")))
```

```r
library(dplyr)

df_agrupado <- df_filtrado %>%
  group_by(DATA_VENDA) %>%
  summarise(Vendas_Totais = sum(QTD_VENDIDA))
```

```r
p <- ggplot(df_agrupado, aes(x = DATA_VENDA, y = Vendas_Totais, fill = as.factor(year(DATA_VENDA)))) +
  geom_bar(stat = "identity", position = "dodge") +
  scale_fill_manual(values = c("blue", "red")) +
  labs(x = "Data de Venda", y = "Vendas Totais", fill = "Ano") +
  theme_minimal() +
  theme(axis.text.x = element_text(angle = 90, hjust = 1))

# Imprime o gráfico no dispositivo R
print(p)

# Salva o gráfico com resolução maior
ggsave("grafico_vendas_altas_resolucao.png", plot = p, width = 10, height = 8, dpi = 300)
```

```r
library(shiny)
library(ggplot2)

# Definindo a UI (User Interface)
ui <- fluidPage(
  titlePanel("Dashboard de Vendas de Medicamentos"),
  sidebarLayout(
    sidebarPanel(
      helpText("Vendas totais de medicamentos controlados e antimicrobianos")
    ),
    mainPanel(
      plotOutput("vendasPlot")  # Output do gráfico
    )
  )
)

# Definindo o servidor
server <- function(input, output) {
  
  # Renderizando o gráfico de barras ggplot2
  output$vendasPlot <- renderPlot({
    ggplot(df_agrupado, aes(x = DATA_VENDA, y = Vendas_Totais, fill = as.factor(format(DATA_VENDA, "%Y")))) +
      geom_bar(stat = "identity", position = "dodge") +
      scale_fill_manual(values = c("blue", "red")) +
      labs(x = "Data de Venda", y = "Vendas Totais", fill = "Ano") +
      theme_minimal() +
      theme(axis.text.x = element_text(angle = 90, hjust = 1))
  })
}

# Executando o aplicativo Shiny
shinyApp(ui = ui, server = server)
```
